```python
from collections import deque

m, n = map(int, input().split())
tomato = []
for i in range(n):
    line = list(map(int, input().split()))
    tomato.append(line)
    
#1: 익은 토마토, 0: 익지 않은 토마토, -1: 토마토가 없는 칸

cnt = [[0]*m for i in range(n)]
for i in range(n):
    for j in range(m):
        if tomato[i][j] == -1: cnt[i][j] = 1
        elif tomato[i][j] == 1: cnt[i][j] = 1
dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]

def bfs(x, y):
    visit = [[0]*m for i in range(n)]
    q = deque()
    q.append([x, y])
    cnt[x][y] = 1
    visit[x][y] = 1
    while q:
        x, y = q[0][0], q[0][1]
        q.popleft()
        for i in range(4):
            nx = x+dx[i]
            ny = y+dy[i]
            if n>nx>=0 and m>ny>=0:
                if tomato[nx][ny] == 0 and visit[nx][ny] == 0:
                    visit[nx][ny] = 1
                    if cnt[nx][ny] == cnt[x][y]+1:
                        return cnt[x][y]+1
                    cnt[nx][ny] = cnt[x][y]+1
                    q.append([nx, ny])
                    
    for i in range(n):
        if cnt[i].count(0) != 0: return -1
    return max(max(cnt))-1


flag = True
ans = 0
for i in range(n):
    for j in range(m):
        if tomato[i][j] == 1:
            ans = bfs(i, j)
        else: flag = False
            
if flag == True: print(0)
else: print(ans)
```

```python
from collections import deque

n, k = map(int, input().split())
Max = 100001
visit = [0]*Max

def bfs(n, k):
    q = deque()
    q.append(n)
    visit[n] = 1
    while q:
        n = q.popleft()
        if n == k: return visit[n]
        dx = [1, -1, n]
        for i in range(3):
            nn = n + dx[i]
            # 범위 안에 있고 방문하지 않았거나 최소 경로로 방문하지 않았을 경우
            if 0<=nn<Max and (visit[nn] == 0 or visit[nn] > visit[n]+1):
                q.append(nn)
                visit[nn] += visit[n]+1
print(bfs(n, k)-1)
```

```python
from collections import deque

n, m = map(int, input().split())
mm = [list(map(int,[*input()])) for k in range(n)]
    
print(mm)
```

```python
visit = [[[0]*m for i in range(n)] for j in range(2)]
dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]
mapp = mm[:]
    
def bfs(y, x): # y:세로, x: 가로
    q = deque()
    q.append([y, x, 0])
    visit[0][y][x] = 1
    visit[1][y][x] = 1
    while q:
        y, x, z = q.popleft()
        if y == n-1 and x == m-1: 
            if visit[0][y][x] == 0: return visit[1][y][x]
            elif visit[1][y][x] == 0: return visit[0][y][x]
            else: return min(visit[0][y][x], visit[1][y][x])
        for i in range(4):
            nx, ny = x + dx[i], y + dy[i]
            if 0<=nx<m and 0<=ny<n:
                if visit[z][ny][nx] == 0 and mapp[ny][nx] == 0:
                    visit[z][ny][nx] = visit[z][y][x]+1
                    q.append([ny,nx, z])
                if z ==0 and visit[z+1][ny][nx] == 0 and mapp[ny][nx] == 1: #벽을 부시기 전일 때
                        visit[z+1][ny][nx] = visit[z][y][x]+1
                        q.append([ny,nx, z+1])
    return -1

print(bfs(0, 0))
```
